# Some Count Plots

```python
# Some Count plots

sns.set()
plt.subplots(figsize=(20,7))
sns.set(style="whitegrid")
ax = sns.countplot(x="department", hue="is_promoted", data=data)
```

```python
# With gender
sns.set()
plt.subplots(figsize=(10,7))
sns.set(style="whitegrid")
ax = sns.countplot(x="gender", hue="is_promoted", data=data)
```

```python
# With education
sns.set()
plt.subplots(figsize=(10,7))
sns.set(style="whitegrid")
ax = sns.countplot(x="education", hue="is_promoted", data=data)
```

##### Below Secondary gets no promotion at all

```python
# With region
sns.set()
plt.subplots(figsize=(20,7))
sns.set(style="whitegrid")
ax = sns.countplot(x="region", hue="is_promoted", data=data,orient='h')
```

```python
# Some Count plots

sns.set()
plt.subplots(figsize=(20,7))
sns.set(style="whitegrid")
ax = sns.countplot(x="previous_year_rating", hue="is_promoted", data=data)
```

Rating plays an important role. 1.0 rating is more or less not promoted at all. 
Same goes with 2.0 

We need to find the % of each category to say if each category has some definitive information.

```python
data[data["is_promoted"]==1]["region"].value_counts()
data[data["is_promoted"]==0]["region"].value_counts()
```

people of every region is promoted but there is definitely a disparity.

```python
sns.FacetGrid(data, hue="is_promoted", height=5) \
   .map(sns.distplot, "avg_training_score") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(data, hue="is_promoted", height=5) \
   .map(sns.distplot, "age") \
   .add_legend();

plt.show();
```

```python
sns.set_style("whitegrid");
sns.pairplot(data, hue="is_promoted",vars=["avg_training_score", "awards_won?", "KPIs_met >80%", "no_of_trainings"])
plt.show()
```

if avg_training_score and no_of_trainings is low. You are promoted

If the number of training points are high. Approximately above 6, The likelihood of getting promoted drop like shit

If you hit KPI's regularly there is a pretty high chance of promotions
